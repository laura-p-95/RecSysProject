```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
URM_training = pd.read_csv("/gdrive/My Drive/Recommender_Sys/recommender-system-2021-challenge-polimi/data_train.csv", 
                                sep=",", 
                                header=None, 
                                skiprows = [0],
                                engine='python')

URM_training.columns = ["UserID", "ItemID", "Interaction"]
#Items are the TVShows!
```

```python
URM_training.head(n=20)
```

```python
URM_train_group = URM_training.groupby(["ItemID"])['Interaction'].sum().sort_values(ascending=False)
```

This line of code orders the Items by the number of interactions. Now we will know which show is the most popular and which shows are instead not that famous.

Here we sum the 'Interaction' column and then sort in a descending way, from the most popular to the least.
